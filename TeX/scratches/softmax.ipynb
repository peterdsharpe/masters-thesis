```python
import aerosandbox as asb
import aerosandbox.numpy as np
from aerosandbox.tools.pretty_plots import plt, sns, mpl, show_plot


x = np.linspace(-4, 8, 200)
f1 = -3 * x - 4
f2 = -0.25 * x + 1
f3 = 2 * x - 12

fig, ax = plt.subplots(1, 1, figsize=(5, 4), dpi=200)
plt.plot(x, f1, label=r"$f_1(x)$", alpha=0.8)
plt.plot(x, f2, label=r"$f_2(x)$", alpha=0.8)
plt.plot(x, f3, label=r"$f_3(x)$", alpha=0.8)
plt.plot(x, np.softmax(f1, f2, f3, hardness=1),
         label=r"softmax$(f_1, f_2, f_3; x)$",
         color="k",
         alpha=0.6,
         )
plt.ylim(-5, 10)

from matplotlib.ticker import MultipleLocator
plt.gca().xaxis.set_major_locator(MultipleLocator(2))
plt.gca().xaxis.set_minor_locator(MultipleLocator(1))
plt.gca().yaxis.set_major_locator(MultipleLocator(2))
plt.gca().yaxis.set_minor_locator(MultipleLocator(1))

show_plot(
    r"Softmax of 3 Linear Functions",
    r"$x$",
    r"$f(x)$",
    show=False
)
plt.savefig(r"C:\Users\User\Dropbox (MIT)\School\Grad School\2021 Spring\Thesis\TeX\figures\softmax.pgf")
plt.show()
```
