```python

import aerosandbox as asb
import aerosandbox.numpy as np
from aerosandbox.tools.pretty_plots import plt, sns, mpl, show_plot

x = np.linspace(-3, 3, 200)
alpha = 1

f = np.abs(x)
f_hat_1 = x * np.tanh(x / alpha)
f_hat_2 = np.sqrt(x ** 2 + alpha ** 2)

fig, ax = plt.subplots(1, 1, figsize=(5,4), dpi=200)
plt.plot(x, f, "gray", label="$f(x) = |x|$")
plt.plot(x, f_hat_1, label = r"$\hat{f}_1(x) = x \tanh(x / \alpha)$")
plt.plot(x, f_hat_2, label = r"$\hat{f}_2(x) = \sqrt{x^2 + \alpha^2}$")

from matplotlib.ticker import MultipleLocator
plt.gca().xaxis.set_major_locator(MultipleLocator(1))
plt.gca().xaxis.set_minor_locator(MultipleLocator(0.5))
plt.gca().yaxis.set_major_locator(MultipleLocator(1))
plt.gca().yaxis.set_minor_locator(MultipleLocator(0.5))


show_plot(
    r"Approximations of $f(x) = |x|$",
    r"$x$",
    r"$\hat{f}(x)$",
    show=False
)
plt.savefig(r"C:\Users\User\Dropbox (MIT)\School\Grad School\2021 Spring\Thesis\TeX\figures\absapprox.pgf")
plt.show()
```
