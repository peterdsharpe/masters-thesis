```python
import aerosandbox as asb
import aerosandbox.numpy as np
from aerosandbox.tools.pretty_plots import plt, sns, mpl, show_plot

x = np.linspace(-1.25, 2.5, 500)


def f(x):
    return np.where(x < 0, -x, (x - 1) ** 2 - 1)


fig, ax = plt.subplots(figsize=(4,3))
plt.plot(x, f(x))
plt.annotate(
    text="Nonconvex\ndiscontinuity\nin $f'(x)$",
    xy=(0, 0),
    xytext=(0.5, 0.5),
    xycoords="data",
    fontsize=8,
    va="center",
    arrowprops={
        "color"     : "gray",
        "width"     : 0.25,
        "headwidth" : 4,
        "headlength": 6,
    }
)

from matplotlib.ticker import MultipleLocator
plt.gca().xaxis.set_major_locator(MultipleLocator(1))
plt.gca().xaxis.set_minor_locator(MultipleLocator(0.5))
plt.gca().yaxis.set_major_locator(MultipleLocator(1))
plt.gca().yaxis.set_minor_locator(MultipleLocator(0.5))

show_plot(
    # r"Example Unimodal Nonconvex Problem",
    xlabel=r"$x$",
    ylabel=r"$f(x)$",
    show=False
)
# plt.savefig(r"C:\Users\User\Dropbox (MIT)\School\Grad School\2021 Spring\Thesis\TeX\figures\nonconvex.pgf")
plt.show()
```
